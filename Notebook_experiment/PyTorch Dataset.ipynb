# Show the PyTorch Dataset for COSMOS

```python
import configparser
import csv
import os
import os.path as ops
import pickle

from PIL import Image
import numpy as np
import scipy
import torch

import matplotlib.pyplot as plt
import argparse
import glob
import json
import cv2

import sys
sys.path.append("..")
from dataset.cosmos_dataset import COSMOSDataset, COSMOSDataset_AUG
```

```python
dataset_path = '../../COSMOS_data_tracktor/detection_dataset/maskrcnn_lablled-bbox-train-val'
dataset_path_aug = '../../COSMOS_data_tracktor/detection_dataset/detection_dataset3'
```

# Define Dataset

```python
dataset = COSMOSDataset(dataset_path)
```

# Show an example from Dataset

```python
dataset[1505]
```

```python
frame_img, frame_ann = dataset[1505]
```

```python
frame_img = np.array(frame_img)
boxes = np.array(frame_ann['boxes'])
labels = np.array(frame_ann['labels'])
```

```python
import matplotlib.patches as patches
plt.figure(figsize=(20,20))
plt.imshow(frame_img)
currentAxis=plt.gca()
for i, label in enumerate(labels):
    if label == 2:
        rect=patches.Rectangle((boxes[i][0], boxes[i][1]),boxes[i][2]-boxes[i][0],boxes[i][3]-boxes[i][1],linewidth=2,edgecolor='b',facecolor='none')
    elif label == 1:
        rect=patches.Rectangle((boxes[i][0], boxes[i][1]),boxes[i][2]-boxes[i][0],boxes[i][3]-boxes[i][1],linewidth=2,edgecolor='r',facecolor='none')
    else:
        continue
    currentAxis.add_patch(rect)
plt.show()
```
